```python
cust = pd.read_csv('customer_data_unsupervised.csv')
cust.head()
```

```python
#converting categorical string to integers
cust['Registration_status'] = cust['Registration_status'].replace(['Active', 'Inactive'],[1,0])
cust['marketing_notification'] = cust['marketing_notification'].replace(['Yes', 'No'],[1,0])
cust.head()
```

```python
#dropping columns that are not needed
data = cust.drop(columns=['first_name','last_name','email','phone','Acquisition_time'])
data.head()
```
